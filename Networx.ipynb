````python
from collections import deque

def bfs(graph, start):

    visited = set()
    queue = deque([start])

    visited.add(start)

    while queue:

        node = queue.popleft()
        print(node, end=" ")

        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)
````

````python
graph = {
    'A': ['B','C'],
    'B': ['D','E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}

bfs(graph, 'A')
````

# NetworkX Python Library

## 1. Introduction

**NetworkX** is a Python library used to create, analyze, and visualize **graphs (networks)**.

A graph contains:

* **Nodes (Vertices)** → Objects
* **Edges (Links)** → Connection between objects

Examples of networks:

* Social networks (Facebook friends)
* Road maps
* Computer networks
* Recommendation systems
* Web links

# 3. Creating Graphs

### Simple Graph

```python
import networkx as nx

G = nx.Graph()
```

### Directed Graph

```python
G = nx.DiGraph()
```

### MultiGraph (multiple edges allowed)

```python
G = nx.MultiGraph()
```

# 4. Adding Nodes

### Add single node

```python
G.add_node(1)
```

### Add multiple nodes

```python
G.add_nodes_from([2,3,4])
```

### Add node with attributes

```python
G.add_node(5, role="teacher")
```

# 5. Adding Edges

### Single edge

```python
G.add_edge(1,2)
```

### Multiple edges

```python
G.add_edges_from([(1,3),(2,4)])
```

### Edge with attributes

```python
G.add_edge(1,5, weight=10)
```

# 6. Graph Information

```python

```
