```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# ------------------ ADJACENCY LIST ------------------

# This is a weighted adjacency list.
# Dictionary structure:
#   Key   → Node
#   Value → List of (Neighbor, Weight)

adj_list = {
    "A": [("B", 5), ("D", 6)],   # A is connected to B (weight 5) and D (weight 6)
    "B": [("A", 5), ("C", 3)],   # B is connected to A (5) and C (3)
    "C": [("B", 3), ("D", 4)],   # C is connected to B (3) and D (4)
    "D": [("C", 4), ("A", 6)]    # D is connected to C (4) and A (6)
}


# ------------------ CREATE GRAPH ------------------

# Create an empty UNDIRECTED graph
# (Edges will work in both directions automatically)
G = nx.Graph()


# ------------------ ADD EDGES FROM ADJ LIST ------------------

# Loop through each node in the adjacency list
for node in adj_list:

    # For each node, get its list of neighbors
    for neighbor, weight in adj_list[node]:

        # Add edge between node and neighbor
        # weight=weight stores the weight as an edge attribute
        G.add_edge(node, neighbor, weight=weight)


# ------------------ DRAW GRAPH ------------------

# spring_layout automatically calculates good positions
# for nodes using force-directed algorithm
pos = nx.spring_layout(G)

# Draw the graph
nx.draw(
    G,                    # Graph object
    pos,                  # Positions of nodes
    with_labels=True,     # Show node names (A, B, C, D)
    node_size=500         # Size of nodes
)


# ------------------ DRAW EDGE WEIGHTS ------------------

# Extract edge weight values from graph
# Returns dictionary like: {('A','B'):5, ('A','D'):6, ...}
edge_labels = nx.get_edge_attributes(G, 'weight')

# Draw weight labels on edges
nx.draw_networkx_edge_labels(
    G,
    pos,
    edge_labels=edge_labels
)


# ------------------ TITLE & SHOW ------------------

# Add title to the graph
plt.title("Weighted Graph (Adjacency List)")

# Display the graph window
plt.show()
```
