```python
import networkx as nx
import matplotlib.pyplot as plt

# Create Graph
G = nx.Graph()

edges = [
    ('A', 'B'), ('A', 'C'),
    ('B', 'D'), ('B', 'E'),
    ('C', 'F')
]

G.add_edges_from(edges)

# DFS Function
def dfs(graph, start, visited=None):
    if visited is None:
        visited = []
    
    visited.append(start)
    
    for neighbor in graph.neighbors(start):
        if neighbor not in visited:
            dfs(graph, neighbor, visited)
    
    return visited

# Run DFS
dfs_result = dfs(G, 'A')
print("DFS Traversal:", " -> ".join(dfs_result))

# Draw Graph
pos = nx.spring_layout(G)  # Layout style
nx.draw(G, pos, with_labels=True, node_size=2000)
plt.title("Graph Visualization (DFS Example)")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt
from collections import deque

# Create Graph
G = nx.Graph()

edges = [
    ('A', 'B'), ('A', 'C'),
    ('B', 'D'), ('B', 'E'),
    ('C', 'F')
]

G.add_edges_from(edges)

# BFS Function
def bfs(graph, start):
    visited = []
    queue = deque([start])

    while queue:
        node = queue.popleft()
        if node not in visited:
            visited.append(node)
            for neighbor in graph.neighbors(node):
                queue.append(neighbor)

    return visited

# Run BFS
bfs_result = bfs(G, 'A')
print("BFS Traversal:", " -> ".join(bfs_result))

# Draw Graph
pos = nx.spring_layout(G)

# Color visited nodes
node_colors = []
for node in G.nodes():
    if node in bfs_result:
        node_colors.append("lightgreen")
    else:
        node_colors.append("lightblue")

nx.draw(G, pos, with_labels=True, node_size=2000, node_color=node_colors)
plt.title("BFS Graph Visualization")
plt.show()
```
